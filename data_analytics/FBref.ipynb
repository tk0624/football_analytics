```python
def generate_scouting_url(player_name, mapping_csv_path, league_id='12524'):
    # mapping.csvを読み込む
    df = pd.read_csv(mapping_csv_path, encoding='latin1')

    # PlayerFBref列を小文字に変換して一致する行を検索
    df['PlayerFBref_lower'] = df['PlayerFBref'].str.lower()
    player_name_lower = player_name.lower()

    # 一致する行を取得
    matched_row = df[df['PlayerFBref_lower'] == player_name_lower]

    if matched_row.empty:
        print(f"Player '{player_name}' not found in mapping.csv.")
        return None

    # UrlFBrefからplayer_idとplayer_slugを抽出
    url = matched_row.iloc[0]['UrlFBref']
    try:
        parts = url.strip().split('/')
        player_id = parts[5]
        player_slug = parts[6]
    except IndexError:
        print(f"Invalid UrlFBref format for player '{player_name}'.")
        return None

    # スカウティングレポートのURLを生成
    scouting_url = f"https://fbref.com/en/players/{player_id}/scout/{league_id}/{player_slug}-Scouting-Report"
    return scouting_url

# 使用例
player_name = 'Kevin De Bruyne'
mapping_csv_path = 'fbref_to_tm_mapping.csv'  # mapping.csvのパスを指定
scouting_url = generate_scouting_url(player_name, mapping_csv_path)

if scouting_url:
    print(f"Scouting URL for {player_name}: {scouting_url}")
```

```python
# ファイルパスの指定
a_csv_path = 'players_wage_ranking_PL.csv'
mapping_csv_path = 'fbref_to_tm_mapping.csv'

# CSVファイルの読み込み
df_a = pd.read_csv(a_csv_path, encoding='latin1')
df_mapping = pd.read_csv(mapping_csv_path, encoding='latin1')

# Player列とPlayerFBref列を小文字に変換して一致させる
df_a['Player_lower'] = df_a['Player'].str.lower()
df_mapping['PlayerFBref_lower'] = df_mapping['PlayerFBref'].str.lower()

# データフレームのマージ
df_merged = pd.merge(df_a, df_mapping, left_on='Player_lower', right_on='PlayerFBref_lower', how='left')

# UrlFBrefからplayer_idとplayer_slugを抽出し、スカウティングレポートのURLを生成
def extract_url(url):
    try:
        parts = url.strip().split('/')
        player_id = parts[5]
        player_slug = parts[6]
        return f'https://fbref.com/en/players/{player_id}/scout/12524/{player_slug}-Scouting-Report'
    except:
        return None

df_merged['24-25_url'] = df_merged['UrlFBref'].apply(extract_url)

# 不要な列の削除
df_merged.drop(['Player_lower', 'PlayerFBref_lower'], axis=1, inplace=True)

# 結果の保存
df_merged.to_csv('a_with_urls.csv', index=False)
```

```python
import requests
from bs4 import BeautifulSoup
import re

url = 'https://fbref.com/en/comps/9/stats/Premier-League-Stats'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
player_ids = set()
for a in soup.find_all('a', href=True):
    match = re.match(r'^/en/players/([a-f0-9]{8})/', a['href'])
    if match:
        player_ids.add(match.group(1))

print(f"取得した選手IDの数: {len(player_ids)}")
print(player_ids)
```
